## Using only the Python Standard Library

## Read in the data

```python
import csv

input_file = '/Users/linali/Desktop/24 Fall/Computing in Context/project 1/Electric_Consumption_And_Cost__2010_-_Feb_2023__20241103.csv'

# open input files
with open(input_file, 'r') as csvfile:

    NYC_electricity = []
    reader = csv.DictReader(csvfile)
    for row in reader:
        NYC_electricity.append(row)
```

## Compute:
1. The mean
1. The median
1. The mode

```python
#calculate mean
consumption = []
for row in NYC_electricity:
    consumption.append(row['Consumption (KWH)'])
    
consumption = [float(row['Consumption (KWH)']) for row in NYC_electricity]
mean_value = sum(consumption)/len(consumption)
print('Mean:', f'{mean_value:.2f}')
```

```python
#calculate median
consumption.sort()
if len(consumption)%2 != 0:
    n = len(consumption)//2
    median_value = consumption[n]
    print('Median:', f'{median_value:.2f}')
else:
    n = len(consumption)//2
    median_value = (consumption[n-1] + consumption[n])/2
    print('Median:', f'{median_value:.2f}')
```

```python
#calculate mode
frequency = {}

for m in consumption:
    if m in frequency:
        frequency[m] += 1
    else:
        frequency[m] = 1

max_frequency = max(frequency.values())
modes = [key for key, count in frequency.items() if count == max_frequency]

for a in modes:
    print ("Mode(s):", f'{a:.2f}')

```

## Data Visualization

```python
consumption_by_borough = {}
for entry in NYC_electricity:
    borough = entry['Borough']
    kwh = float(entry['Consumption (KWH)'])
    
    if borough in consumption_by_borough:
        consumption_by_borough[borough] += kwh
    else:
        consumption_by_borough[borough] = kwh

print(consumption_by_borough)
```

```python
max_consumption = max(consumption_by_borough.values())
bar_character = '█'
max_bar_length = 50
for borough, kwh in consumption_by_borough.items():
    bar_length = int((kwh/max_consumption) * max_bar_length)
    bar = bar_character * bar_length
    print(f'{borough:25}:{bar}{kwh:,.0f}')
```
